Question 1: What is hypothesis testing in statistics?

Answer:
Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It helps determine whether there is enough evidence to reject a null hypothesis in favor of an alternative hypothesis. In essence, it evaluates claims or assumptions made about a dataset using probability and statistical analysis.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:

Null Hypothesis (H₀): It is a statement that assumes there is no significant difference or effect in the population. It represents the status quo or default assumption.
Example: H₀: μ = 50 (The population mean is 50).

Alternative Hypothesis (H₁ or Ha): It is a statement that contradicts the null hypothesis. It suggests that there is a significant difference or effect.
Example: H₁: μ ≠ 50 (The population mean is not 50).

Difference:
The null hypothesis assumes no change or relationship, while the alternative hypothesis suggests the presence of a change or relationship.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It defines the threshold for statistical significance. Commonly used levels are 0.05, 0.01, and 0.10.

If the p-value ≤ α, we reject the null hypothesis (statistically significant).
If the p-value > α, we fail to reject the null hypothesis.

Thus, α acts as a cutoff that determines how strong the evidence must be to reject H₀.

Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

Type I Error (α): Occurs when the null hypothesis is true but is rejected.
Example: A medical test wrongly indicates that a healthy person has a disease.

Type II Error (β): Occurs when the null hypothesis is false but is not rejected.
Example: A medical test fails to detect a disease in a person who actually has it.

In summary:

Type I → False Positive

Type II → False Negative

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:

Aspect	             Z-Test	                      T-Test
Population Variance	Known	Unknown
Sample Size	Large (n > 30)	Small (n ≤ 30)
Distribution Used	Standard Normal (Z)	Student’s t-distribution
Use Case	Comparing sample mean to population mean when σ is known	Comparing means when σ is unknown

Example Use:

Use Z-test for large samples with known variance.

Use T-test for small samples or when variance is unknown.

```python
# Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

# Answer:

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()

# Output:
# A histogram showing a symmetric distribution centered around 5 (since p=0.5).
```

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

# Answer:

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis: H0: μ = 50
mu = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", sample_mean)
print("Z-value:", z)
print("P-value:", p_value)

# Interpretation:
# If p-value > 0.05, we fail to reject the null hypothesis.
# The result likely shows no significant difference from μ = 50, meaning the sample mean is statistically close to 50.
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

# Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate normal distribution
data = np.random.normal(loc=50, scale=5, size=100)
mean = np.mean(data)
std_err = stats.sem(data)

# 95% confidence interval
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print("Mean:", mean)
print("95% Confidence Interval:", conf_interval)

# Plot
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean:.2f}')
plt.title("Normal Distribution Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Interpretation:
# The 95% confidence interval shows the range within which the true population mean likely lies.
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
# Answer:
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_zscores(data):
    z_scores = stats.zscore(data)
    return z_scores

# Example dataset
data = np.random.normal(100, 15, 100)
z_scores = calculate_zscores(data)

print("First 5 Z-scores:", z_scores[:5])

# Plot histogram
plt.hist(z_scores, bins=15, edgecolor='black', color='lightgreen')
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

```
